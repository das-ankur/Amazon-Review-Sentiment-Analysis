## **Import Libraries**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import math
import warnings
warnings.filterwarnings('ignore') # Hides warning
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings("ignore",category=UserWarning)
sns.set_style("whitegrid") # Plotting style
np.random.seed(7) # seeding random number generator
```

## **Split Dataset**

```python
from sklearn.model_selection import StratifiedShuffleSplit
print("Size of the data before : ",len(df))
df = df.dropna(subset=["reviews.rating"])
# Removes all NAN in reviews.rating
print("After of the data beforr : ",len(df))
df["reviews.rating"] = df["reviews.rating"].astype(int)
split = StratifiedShuffleSplit(n_splits=5, test_size=0.2)
for train_index, test_index in split.split(df, df["reviews.rating"]):
    train_data = df.reindex(train_index)
    test_data = df.reindex(test_index)
```

```python
print("Size of train data : ",len(train_data))
print("Size of test data : ",len(test_data))
print("Distribution of reviews :",test_data["reviews.rating"].value_counts()/len(test_data), sep = "\n")
```

## **Data Exploration**

```python
reviews = train_data.copy()
reviews.head()
```

```python
print("No. of unique reiviewer : ",len(reviews["name"].unique()))
print("No. of unique asins : ",len(reviews["asins"].unique()))
print(reviews.info())
print("Groupby 'asins' : ")
print(reviews.groupby("asins")["name"].unique())
```

```python
fig = plt.figure(figsize=(16,10))
ax1 = plt.subplot(211)
ax2 = plt.subplot(212, sharex = ax1)
reviews["asins"].value_counts().plot(kind="bar", ax=ax1, title="ASIN Frequency")
np.log10(reviews["asins"].value_counts()).plot(kind="bar", ax=ax2, title="ASIN Frequency (Log10 Adjusted)")
plt.show()
```

```python
print("Average rating : ", reviews["reviews.rating"].mean())
asins_count_ix = reviews["asins"].value_counts().index
plt.subplots(2,1,figsize=(16,12))
plt.subplot(2,1,1)
reviews["asins"].value_counts().plot(kind="bar", title="ASIN Frequency")
plt.subplot(2,1,2)
sns.pointplot(x="asins", y="reviews.rating", order=asins_count_ix, data=reviews)
plt.xticks(rotation=90)
plt.show()
```

```python
def sentiments(rating):
    if (rating == 5) or (rating == 4):
        return "Positive"
    elif rating == 3:
        return "Neutral"
    elif (rating == 2) or (rating == 1):
        return "Negative"
# Add sentiments to the data
train_data["Sentiment"] = train_data["reviews.rating"].apply(sentiments)
test_data["Sentiment"] = test_data["reviews.rating"].apply(sentiments)
print(train_data["Sentiment"][:20])
```
